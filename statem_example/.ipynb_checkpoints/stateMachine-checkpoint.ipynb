## This example presents a build `on-the-fly` method for creating a pythonic state machne lanaguage. This code is based on the examples provied by textX here --> http://textx.github.io/textX/stable/tutorials/state_machine/

```python
import os
```

```python
# Review the language discrption for the parser. This is typically some keywords for the parser to find... like `Package` or 'Part' 
from textx import metamodel_from_file
from textx.export import metamodel_export
entity_mm = metamodel_from_file('state_machine.tx')
metamodel_export(entity_mm, 'state_machine.dot')
```

```python
# Now plot the state machine meta-model
import graphviz
with open("state_machine.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# a prgramatic way to build a graph of a model... this model is an example of a state machine
from textx.export import model_export
model = entity_mm.model_from_file('miss_grant_controller.sm')
model_export(model, 'miss_grant_controller.sm.dot')
```

```python
# vizualiaztion of the python object graph (from parsing model... creates a generic view)
import graphviz
with open("miss_grant_controller.sm.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# translate object graph to state machine dot file
%run smdot.py miss_grant_controller.sm
```

```python
# state machone model
import graphviz
with open("miss_grant_controller.sm.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# run the state machine
%run state_machine.py miss_grant_controller.sm
```
